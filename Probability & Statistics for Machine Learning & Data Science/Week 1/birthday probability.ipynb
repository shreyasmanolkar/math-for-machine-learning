The birthday problem

```python
import matplotlib.pyplot as plt

each_value = []
probabilities = []
probability = 1

for value in range(1, 51):
    each_value.append((365 - value) / 365)
    probability *= each_value[-1]
    probabilities.append(probability)

plt.figure(figsize=(14, 10))

plt.plot(range(1,51), probabilities, marker='.', label='Probability')
plt.axhline(y=0.5, color='r', linestyle='--', label='0.5 Probability')

plt.title('Probability of Shared Birthdays')
plt.xlabel('Number of People')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f"Number of person: 01 and Probability: 1.000")
for idx, value in enumerate(probabilities):
    if idx < 8:
        print(f"Number of person: 0{idx+2} and Probability: {round(value,3)}")
    else:
        print(f"Number of person: {idx+2} and Probability: {round(value,3)}")
```
